##**Funciones**

```python
#Remover acentos y caracteres especiales
def quitar_acento(string):
    return unidecode.unidecode(string)
```

```python
#Comprobar si los elementos dentro de una lista o las columnas de un DF están en otro DF
def comprobar_columnas(lista1, lista2):
  if len(lista1) == len(lista2):
    print('Los DataFrames tienen la misma cantidad de columnas.')
  else:
    print('Los DataFrames no tienen la misma cantidad de columnas.')
  
  not_in = []
   
  for i in lista1:
    if i not in lista2:
      not_in.append(i)

  for i in lista2:
    if i not in lista1:
      not_in.append(i)

  if len(not_in) == 0:
    print('Los DateFrames tienen las mismas columnas. \n')
  else:
    print(f'Las siguientes columnas no están en alguno de los DataFrames: {not_in} \n')
```

```python
#Añadir numero según el día de la semana
def num_dia_sem(fila):
  if fila == 'Lunes':
    v = 1
  elif fila == 'Martes':
    v = 2
  elif fila == 'Miércoles':
    v = 3
  elif fila == 'Jueves':
    v = 4
  elif fila == 'Viernes':
    v = 5
  elif fila == 'Sábado':
    v = 6
  else:
    v = 7
  return v
```
